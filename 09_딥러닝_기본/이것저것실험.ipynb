# autograd(ìë™ë¯¸ë¶„)
- ìë™ ë¯¸ë¶„ì„ ì´ìš©í•´ gradient(ë¯¸ë¶„ê³„ìˆ˜)ë¥¼ ê³„ì‚°í•˜ëŠ” pytorch system.
- ë”¥ëŸ¬ë‹ ëª¨ë¸ì—ì„œ weightì™€ bias tensorë“¤(Parameter)ì€ backpropagation(ì—­ì „íŒŒ)ë¥¼ ì´ìš©í•´ gradientë¥¼ êµ¬í•´ì„œ lossê°€ ì¤„ì–´ë“œëŠ” ë°©í–¥ìœ¼ë¡œ updateë¥¼ í•˜ê²Œëœë‹¤.
- pytorchëŠ” ì´ëŸ° ë¯¸ë¶„ ìˆ˜í–‰ì„ ìë™ìœ¼ë¡œ ì²˜ë¦¬í•´ ì¤€ë‹¤.
    - gradient(ê¸°ìš¸ê¸°)ë¥¼ êµ¬í•œë‹¤ëŠ” ê²ƒì€ ë¯¸ë¶„ì„ í•œë‹¤ëŠ” ê²ƒì„ ë§í•œë‹¤.
- tensorê°€ ë¯¸ë¶„ ê°€ëŠ¥í•˜ë ¤ë©´(gradient ê³„ì‚° ëŒ€ìƒ ë³€ìˆ˜) `requires_grad=True` ë¡œ ì„¤ì •ë˜ ìˆì–´ì•¼ í•œë‹¤. (default: False)

> - **ë¯¸ë¶„**
>   -  (ìˆœê°„) ë³€í™”ìœ¨ì„ ê³„ì‚°í•œë‹¤.
>   - $$\frac{\partial y}{\partial x}$$
>       - xì— ëŒ€í•œ yì˜ ë³€í™”ìœ¨. ì¦‰ xê°€ ë³€í•˜ë©´ yëŠ” ì–¼ë§ˆë‚˜ ë³€í• ê¹Œ?

````python
# 1. ì…ë ¥ê°’ì„ í…ì„œë¡œ ì •ì˜
x = torch.tensor(2.0, requires_grad=True)  # requires_grad=True: ë¯¸ë¶„ì„ ê³„ì‚°í•  ìˆ˜ ìˆë„ë¡ ì„¤ì •

# 2. ì—°ì‚°ì„ ì •ì˜ (y = x^2 + 3x)
y = x**2 + 3*x

# 3. ë¯¸ë¶„ì„ ê³„ì‚° (yì— ëŒ€í•œ xì˜ ë¯¸ë¶„)
y.backward()  # ìë™ìœ¼ë¡œ ì—­ì „íŒŒë¥¼ ê³„ì‚°í•˜ì—¬ gradient ê°’ì„ ê³„ì‚°

# 4. ë¯¸ë¶„ê°’ì„ ì¶œë ¥
print(f"y = x^2 + 3x ì— ëŒ€í•œ xì˜ ë¯¸ë¶„ ê°’: {x.grad}")
````

````python
# ë‹¤ë³€ìˆ˜ í•¨ìˆ˜ì˜ ë¯¸ë¶„
# 1. ë‘ ê°œì˜ ì…ë ¥ê°’ì„ í…ì„œë¡œ ì •ì˜
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

# 2. ë‘ ë³€ìˆ˜ì— ëŒ€í•œ í•¨ìˆ˜ ì •ì˜ (z = x^2 + y^3)
z = x**2 + y**3

# 3. ë¯¸ë¶„ì„ ê³„ì‚° (zì— ëŒ€í•œ xì™€ yì˜ ë¯¸ë¶„)
z.backward()

# 4. xì™€ yì— ëŒ€í•œ ê¸°ìš¸ê¸° ì¶œë ¥
print(f"xì— ëŒ€í•œ ê¸°ìš¸ê¸°: {x.grad}")  # dz/dx
print(f"yì— ëŒ€í•œ ê¸°ìš¸ê¸°: {y.grad}")  # dz/dy
````
